#Weiner Process

```python
import numpy as np
import yfinance as yf
import numpy.random as npr
import matplotlib.pyplot as plt

def weiner_process(dt=0.1, x0=0, n=1000):
  W = np.zeros(n+1)
  t = np.linspace(x0, n, n+1)

  W[1:n+1] = np.cumsum(npr.normal(x0, np.sqrt(dt), n))

  return t, W

def plot_process(t, W):
  plt.figure(figsize=(20, 10))
  plt.plot(t, W)
  plt.xlabel('Time(t)', fontsize=15)
  plt.ylabel('Wiener-process W(t)', fontsize=15)
  plt.title('Wiener Process', fontsize=20)
  plt.show()


if __name__ == '__main__':
  time, data = weiner_process()
  plot_process(time, data)
```

`In Weiner Process it is evident that the stock price may be found negative which is surreal. Not Possible in real life. That's why we will go through the geometric brownian process to get better result with random simulation.`

#Geometric Random Walk

```python
def simulate_geometric_random_walk(S0, T=2, N=1000, mu=1, sigma=0.5):
  dt = T/N
  t = np.linspace(0, T, N)
  #print(dt)
  #print(t)
  W = npr.standard_normal(size=N)
  #dW = W * dt
  W = np.cumsum(W) * np.sqrt(dt)
  X = (mu - 0.5 * sigma**2) * t + sigma * W
  S = S0 * np.exp(X)

  return t, S

def plot_process(t, S):
  plt.figure(figsize=(20, 10))
  plt.plot(t, S)
  plt.xlabel('Time(t)', fontsize=15)
  plt.ylabel('Stock Price S(t)', fontsize=15)
  plt.title('Geometric Browinian Motion', fontsize=20)
  plt.show()


if __name__ == '__main__':
  time, data = simulate_geometric_random_walk(10)
  plot_process(time, data)
```
